```python
df['Amount'] = df['Quantity']*df['Price']
rfm_m = df.groupby('CustomerNo')['Amount'].sum()
rfm_m = rfm_m.reset_index()
rfm_m.head()
```

```python
sns.histplot(rfm_m['Amount'])
```

```python
score = []
for num_clusters in range(1,10):
    kmeans = KMeans(n_clusters=num_clusters, max_iter=300)
    kmeans.fit(rfm_m)
    score.append(kmeans.inertia_)
plt.plot(score)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(rfm_m[['Amount']])
rfm_m['AmountCluster'] = kmeans.predict(rfm_m[['Amount']])

#function for ordering cluster numbers
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final

rfm_m = order_cluster('AmountCluster', 'Amount',rfm_m,False)
rfm_m.groupby('AmountCluster')['Amount'].describe()
```

```python
plot_data = [
    go.Histogram(
        x=rfm_m['Amount']
 )
]
plot_layout = go.Layout(
        title='Amount'
 )
fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```

```python
rfm_f = df.groupby('CustomerNo')['TransactionNo'].count()
rfm_f = rfm_f.reset_index()
rfm_f.columns = ['CustomerNo', 'Frequency']
rfm_f.head()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(rfm_f[['Frequency']])
rfm_f['FrequencyCluster'] = kmeans.predict(rfm_f[['Frequency']])

#function for ordering cluster numbers
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final

rfm_f = order_cluster('FrequencyCluster', 'Frequency',rfm_f,False)
rfm_f.groupby('FrequencyCluster')['Frequency'].describe()
```

```python
plot_data = [
    go.Histogram(
        x=rfm_f['Frequency']
 )
]
plot_layout = go.Layout(
        title='Frequency'
 )
fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```

```python
rfm = pd.merge(rfm_m, rfm_f, on='CustomerNo', how='inner')
rfm.head()
```

```python
maxdate = max(df['Date'])
maxdate
```

```python
mindate = min(df['Date'])
mindate
```

```python
df['days_ago'] = maxdate - df['Date']
df.head()
```

```python
rfm_r = df.groupby('CustomerNo')['days_ago'].min()
rfm_r = rfm_r.reset_index()
rfm_r['Recency'] = rfm_r['days_ago'].dt.days
rfm_r = rfm_r.drop('days_ago',axis = 1)
rfm_r.head()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(rfm_r[['Recency']])
rfm_r['RecencyCluster'] = kmeans.predict(rfm_r[['Recency']])

#function for ordering cluster numbers
def order_cluster(cluster_field_name, target_field_name,df,ascending):
    new_cluster_field_name = 'new_' + cluster_field_name
    df_new = df.groupby(cluster_field_name)[target_field_name].mean().reset_index()
    df_new = df_new.sort_values(by=target_field_name,ascending=ascending).reset_index(drop=True)
    df_new['index'] = df_new.index
    df_final = pd.merge(df,df_new[[cluster_field_name,'index']], on=cluster_field_name)
    df_final = df_final.drop([cluster_field_name],axis=1)
    df_final = df_final.rename(columns={"index":cluster_field_name})
    return df_final

rfm_r = order_cluster('RecencyCluster', 'Recency',rfm_r,False)
rfm_r.groupby('RecencyCluster')['Recency'].describe()
```

```python
plot_data = [
    go.Histogram(
        x=rfm_r['Recency']
 )
]
plot_layout = go.Layout(
        title='Recency'
 )
fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```

```python
rfm = pd.merge(rfm, rfm_r, on='CustomerNo', how='inner')
# rfm.columns = ['CustomerNo', 'Amount','AmountCluster', 'Frequency','FrequencyCluster', 'Recency','RecencyCluster']
rfm.head()
```

```python
attributes = ['Amount','Frequency','Recency']
plt.rcParams['figure.figsize'] = [10,8]
sns.boxplot(data = rfm[attributes], orient="v", palette="Set2" ,whis=1.5,saturation=1, width=0.7)
plt.title("Outliers Variable Distribution", fontsize = 14, fontweight = 'bold')
plt.ylabel("Range", fontweight = 'bold')
plt.xlabel("Attributes", fontweight = 'bold')
```

```python
Q1 = rfm.Amount.quantile(0.05)
Q3 = rfm.Amount.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Amount >= Q1 - 1.5*IQR) & (rfm.Amount <= Q3 + 1.5*IQR)]

Q1 = rfm.Recency.quantile(0.05)
Q3 = rfm.Recency.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Recency >= Q1 - 1.5*IQR) & (rfm.Recency <= Q3 + 1.5*IQR)]

Q1 = rfm.Frequency.quantile(0.05)
Q3 = rfm.Frequency.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Frequency >= Q1 - 1.5*IQR) & (rfm.Frequency <= Q3 + 1.5*IQR)]
```

```python
rfm.head()
```

```python
#calculate overall score and use mean() to see details
rfm['OverallScore'] = rfm['RecencyCluster'] + rfm['FrequencyCluster'] + rfm['AmountCluster']
rfm = order_cluster('OverallScore', 'Recency',rfm,True)
rfm.groupby('OverallScore')['Recency','Frequency','Amount'].mean()
```

```python
rfm['Segment'] = 'Low-Value'
rfm.loc[rfm['OverallScore']>0,'Segment'] = 'Mid-Value' 
rfm.loc[rfm['OverallScore']>1,'Segment'] = 'High-Value' 
```
